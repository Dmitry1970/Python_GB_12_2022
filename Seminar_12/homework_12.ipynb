```python
# Задача 44: В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies?

import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
print(data)
```

```python
# Как бы получилось с get_dummies
pd.get_dummies(data['whoAmI'])
```

```python
# с помощью библиотеки sklearn

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore')
encoder_data = pd.DataFrame(encoder.fit_transform(data[['whoAmI']]).toarray())
final_data = data.join(encoder_data)
print(final_data)
```

```python
final_data.drop('whoAmI', axis= 1, inplace= True)
final_data.columns = ['human', 'robot']
print(final_data)
```
